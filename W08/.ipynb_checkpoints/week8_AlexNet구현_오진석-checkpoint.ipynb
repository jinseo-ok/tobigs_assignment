```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
from keras_radam.training import RAdamOptimizer
from tensorflow.keras.regularizers import l2
from tensorflow.keras import backend as K
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Layer
```

# AlexNet
###### 출처 : https://bskyvision.com/421
   -  AlexNet은 2012년에 개최된 ILSVRC(ImageNet Large Scale Visual Recognition Challenge) 대회의 우승을 차지한 컨볼루션 신경망(CNN) 구조

#### 1) 구조
   - AlexNet의 기본구조는 LeNet-5와 크게 다르지 않다. 2개의 GPU로 병렬연산을 수행하기 위해서 병렬적인 구조로 설계되었다는 점이 가장 큰 변화
   - 총 8개의 레이어 중, 5개의 convolution layers 와 3개의 fully-connected layers로 구성
   - fully_connected layers 에서는 1000개의 카테고리로 분류하기 위해 활성 함수로 softmax 사용
   - 총 약 6천만개의 parameters가 훈련


image.png

#### 2) 구현

```python
model = Sequential()
model
```

```python
inputs = (227, 227, 3)
# 첫번째 layer
model.add(Conv2D(96, (11, 11), strides=4,padding='same',  input_shape=inputs))
model.add(MaxPooling2D(pool_size=(3, 3), strides=2))
# model.add(LocalResponseNormalization(input_shape=model.output_shape[1:]))
# 수렴 속도를 높이기 위해 local response normalization이 시행된다. local response normalization은 특성맵의 차원을 변화시키지 않으므로, 특성맵의 크기는 27 x 27 x 96으로 유지
# 두번째 layer
model.add(Conv2D(256, (5, 5), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(3, 3), strides=2))
# model.add(LocalResponseNormalization(input_shape=model.output_shape[1:]))
# 세번째 layer
model.add(Conv2D(384, (3, 3), activation='relu', padding='same'))
# 네번째 layer
model.add(Conv2D(384, (3, 3), activation='relu', padding='same'))
# 다섯번째 layer
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(3, 3), strides=2))
# 여섯번째 layer
model.add(Flatten())
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
# 일곱번째 layer
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))
# 여덟번째 layer
model.add(Dense(1000, activation='softmax'))


model.compile(optimizer = RAdamOptimizer(learning_rate=1e-3), loss="sparse_categorical_crossentropy", metrics=["acc"])
model.summary()
```
