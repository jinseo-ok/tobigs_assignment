1-5) 마지막으로 고유 벡터 축으로 값을 변환해 볼게요

함수로 한번 정의해 보았어요

https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html

```python
new = X_std.dot(eigenvectors.T[0])
# new = np.concatenate((new, X),axis=0)
new
```

```python
eigenvectors.T[0]
```

```python
def new_coordinates(X,eigenvectors):
    for i in range(eigenvectors.shape[0]):
        if i == 0:
            new = [X_std.dot(eigenvectors.T[i])] # X -> X_std 로 수정
        else:
            new = np.concatenate((new,[X_std.dot(eigenvectors.T[i])]),axis=0)
    return new.T

# 모든 고유 벡터 축으로 데이터를 projection한 값입니다
```

# 2) PCA를 구현해 보세요

위의 과정을 이해하셨다면 충분히 하실 수 있을거에요

```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):
    scaler = StandardScaler()
    x_std = scaler.fit_transform(X) # 스케일링
    features = x_std.T # 전치행렬
    cov_matrix = np.cov(features) # 공분산 행렬 생성
    
    eigenvalues = lin.eig(cov_matrix)[0]
    eigenvectors = lin.eig(cov_matrix)[1]
    
    new_coordinates(x_std,eigenvectors)
    
    new_coordinate = new_coordinates(x_std,eigenvectors)
    
    index = eigenvalues.argsort()
    index = list(index)
    
    for i in range(number):
        if i==0:
            new = [new_coordinate[:,index.index(i)]]
        else:
            new = np.concatenate(([new_coordinate[:,index.index(i)]],new),axis=0)
    return new.T
```

```python
MYPCA(X,3)
```

```python
# 답안
MYPCA(X,3)

# 새로운 축으로 잘 변환되어서 나타나나요?
# 위에서 했던 PCA랑은 차이가 있을 수 있어요 왜냐하면 위에서는 고유값이 큰 축 순서로 정렬을 안했었거든요
```

### 적절한 주성분 개수 결정하기
   - 누적설명률

```python
# 전체 feature 로 PCA 를 진행
pca = PCA()
pca_train = pca.fit_transform(X_train)
features = range(pca.n_components_) # components_ 개수
```

```python
# 주성분이 이루는 축에 따른 분산의 비율 분포
# 100번 째 이하의 주성분 축이 그 이후의 주성분 축에 비해 많은 양을 가지고 있음
# 특히 10번 째 주성분 축쯤에서 급격히 감소함
plt.figure(figsize=(15,8))
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.axvline(x=10, c='r')
plt.show()
```

```python
# 적절한 차원 선택하기
# 누적된 분산의 비율(누적설명률)을 기준으로 주성분 개수 결정
# 43 까지의 주성분이 전체 데이터의 80%의 설명력을 보임
cumsum = np.cumsum(pca.explained_variance_ratio_)
k = np.argmax(cumsum >= 0.8) + 1
print('선택할 차원 수 :', k)
```

```python
k = np.argmax(cumsum >= 0.95) + 1
print('선택할 차원 수 :', k)
```

   - Kaiser’s Rule 활용
 

```python
# Kaiser’s Rule 을 활용하여 고유값이 1 이상으로 이루어지는 주성분들 구하기
# 주성분들의 각 고유값의 최소값이 1 이하로 떨어지는 구간을 확인
length = np.arange(0,765,15)
eigvalues_list = list()

for i in range(1,51): # 너무 오래걸려서 15 개 텀으로 진행
    i = i * 15
    print(i)
    if i < 750:
        pca = PCA(n_components = i)
        pca.fit(X_train)
        eigvalues_list.append(pca.explained_variance_.min()) # 보유하고 있는 주성분의 최소값
    else:
        i = 748
        pca = PCA(n_components = i)
        pca.fit(X_train)
        eigvalues_list.append(pca.explained_variance_.min())
```

```python
# 빨간선은 누석설명률에서 구한 43개의 차원 

x_length = np.arange(15,765,15)
plt.figure(figsize=(15,8))
plt.plot(x_length,eigvalues_list, linewidth = 3)
plt.xlabel('Factor Number')
plt.ylabel('eigenvalue')
plt.axvline(x=43, c='r')
```

```python
# 고유값이 1 이하로 떨어지는 구간은 약 645 차원으로 이루어진 구간
# 누적설명률에서는 43개가 나왔는데 여기서는 645 언저리가 나온 것으로 보아 잘못 이해한 것으로 판단됨
k = np.argmax(np.array(eigvalues_list) < 1)
print('고유값이 1이하로 떨어지는 구간은', k*15)
```
